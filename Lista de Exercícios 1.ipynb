# Equalização de Histograma

```python
import decimal

rk = [0,1,2,3]
nk = [100,146,80,74]
MN = 400
L = 3
prrj = []



def pr_rk(nk,MN):
    x = float(nk/MN)
    #print('x',x)
    return x

cont = 0
print('rk  |  nk  |  pr(rk)')
for r in rk:
    
    prrj.append(pr_rk(nk[cont], MN))
    print(r,'  |  ',nk[cont],'  |  ',prrj[cont])
    cont = cont + 1

print('\n')
cont2 = 0
for r in rk:
    print('p(r',cont2,')= ', nk[cont2],'/', MN,' = ',prrj[cont2])
    cont2 = cont2 + 1
    
print('prrj',prrj)
sk = []
for r in rk:
    sum = 0
    print('\nR', r)
    cont = 0
    while cont <= r:
        sum =  (sum + L * prrj[cont])
        #print('prrj',prrj[cont])
       
        
        print('sum=','(',sum,'+ (',L,'*',prrj[cont],')) =', sum)
        cont += 1
    #int(sum)
    sk.append(sum)
    print('S',r,'= ',round(sum))
```

# Especificação de Histogramas

```python
Zq = [0,1,2,3]
pzzq = [0.1,0.25,0.4,0.25]
L = 4
Gzq = []
def G(Zq, pzzq):
    for z in Zq:
        sum = 0.00
        i = 0
       
        appendprint = []
        appendprint.append('G(Z')
        appendprint.append(z)
        appendprint.append(') = ')
        print('\n\nG(Z',z,')= ')
        while i <= z:
            appendprint.append(L-1)
            appendprint.append('*')
            appendprint.append((pzzq[i]))
            appendprint.append('=')
            appendprint.append((L-1) * (pzzq[i]))
            appendprint.append('+')
            sum  = sum + ((L-1) * (pzzq[i])) 
            print('(L-1) * (',pzzq[i],') = ',(L-1) * (pzzq[i]),'+')
            i += 1
        print('=================== sk',(sum),'= Zq', z)
        #print(appendprint)
        Gzq.append(round(sum))
    return Gzq

print('G(Zq)',G(Zq, pzzq))
print(Gzq)
print('\n\nsk | zq')

print(sk)
for i in range(len(sk)):
    #
    
    #print(sk[i])
    print(round(sk[i]) ,' || ',Gzq[i],'|', Zq[i])
```
